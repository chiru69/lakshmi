```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('mtcars.csv')
plt.hist(df['mpg'], bins=30, edgecolor='black')
plt.xlabel('mpg')
plt.ylabel('frequency')
plt.title('histogram')
plt.show()
```

```python
plt.scatter(df['wt'], df['mpg'])
plt.xlabel('weight of the car')
plt.ylabel('mpg')
plt.title('scatter plot')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
transmission_type=df['disp'].value_counts()
transmission_type.plot(kind='bar')
plt.xlabel('transmission_typer')
plt.ylabel('frequency')
plt.title('frequency distribution')
plt.show()
```

```python
plt.boxplot(df['mpg'])
plt.ylabel('mpg')
plt.title('box plot')
plt.show()
```

```python
import pandas as pd
pd.set_option('display.max_rows',None)
df=pd.DataFrame({
    'book_name':['Book1','Book2','Book3','Book4','Book1','Book2','Book3','Book5'],
    'book_type':['math','physics','computer','science','math','physics','computer','english'],
    'book_id':[1,2,3,4,1,2,3,5]
})
print("column with count from GroupBy")
result=df.groupby(['book_name','book_type'])['book_type'].count().reset_index(name='count')
print(result)
```
